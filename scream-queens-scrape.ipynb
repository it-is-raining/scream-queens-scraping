```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Scream_queen"

html = requests.get(url).text
```

```python
soup = BeautifulSoup(html)
type(soup)
```

```python
soup.title
```

```python
scream_queens = soup.select(".wikitable")

len(scream_queens)
```

```python
row_els = soup.select(".wikitable tr")

row_els
```

```python
header_cells = soup.select(".wikitable th")
scream_queen_headers = [ header.text.strip() for header in header_cells ]

scream_queen_headers
```

```python
scream_queen_entries = []

for row in row_els:
    row_cells = []
    for cell in row.select("td"):
        row_cells.append(cell.text.strip())
    scream_queen_entries.append(row_cells)

scream_queen_entries
```

```python
# df = pd.DataFrame(row_els)

# df

import pandas as pd
scream_df = pd.DataFrame(scream_queen_entries, columns=scream_queen_headers)
scream_df
```
